```python
import torch
from torch.utils.data import dataloader
from torchvision import datasets, transforms

```

```python
# Transforming the data with torchvision transform

data_transform = transforms.Compose([
    transforms.Resize(size=(64, 64)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ToTensor()
])
```

### Visulizing Transforms

```python
def plot_image_path(image_path, transform, n=3, seed=None):
    if seed:
        random.seed(seed)
    random_image_path = random.sample(image_path, k=n)
    for image_path in random_image_path:
        with Image.open(image_path) as f:
            fig, ax = plt.subplots(nrows=1, ncols=2)
            ax[0].imshow(f)
            ax[0].set_title(f"/n Original Image size: {f.size}")
            ax[0].axis(False)

            transform_image = transform(f).permute(1,2,0)
            ax[1].imshow(transform_image)
            ax[1].set_title(f"/n Transformed Image size: {transform_image.size}")
            ax[1].axis("off")


            fig.suptitle(f"/n Class {image_path.parent.stem}", fontsize=16)

plot_image_path(image_path=image_path_list,
                transform=data_transform,
                n=3,
                seed=42)
```
